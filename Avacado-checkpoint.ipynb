##### Treating missing values by replacing them with mean

Before that the unnamed,Date and year column does not give any information so dropping that column

```python
data_new=data.drop(['Unnamed: 0','Date','year'],axis=1)
```

```python
data_new.head()
```

```python
data_new["AveragePrice"]=data_new["AveragePrice"].fillna(np.mean(data_new["AveragePrice"]))
data_new["Total Volume"]=data_new["Total Volume"].fillna(np.mean(data_new["Total Volume"]))
data_new["4046"]=data_new["4046"].fillna(np.mean(data_new["4046"]))
data_new["4225"]=data_new["4225"].fillna(np.mean(data_new["4225"]))
data_new["4770"]=data_new["4770"].fillna(np.mean(data_new["4770"]))
data_new["Total Bags"]=data_new["Total Bags"].fillna(np.mean(data_new["Total Bags"]))
data_new["Small Bags"]=data_new["Small Bags"].fillna(np.mean(data_new["Small Bags"]))
data_new["Large Bags"]=data_new["Large Bags"].fillna(np.mean(data_new["Large Bags"]))
data_new["XLarge Bags"]=data_new["XLarge Bags"].fillna(np.mean(data_new["XLarge Bags"]))
```

For all the continuous columns replaced the missing values with mean.Let us check for confirmation

```python
data_new.isnull().sum()
```

For all the categorical columns need to replace the missing values for this we need to do encoding and then replace the missing values.

```python
data_new.dtypes
```

Let us fill the missing values of categorical data

```python
data_new['type'].value_counts()
```

```python
data_new['type'].fillna(data_new['type'].value_counts().index[0],inplace=True)
data_new['region'].fillna(data_new['region'].value_counts().index[0],inplace=True)
```

```python
data_new.isna().sum()
```

so all the missing values has been handled so let us know encode our data frame

```python
from sklearn.preprocessing import OrdinalEncoder
enc=OrdinalEncoder()
```

```python
for i in data_new.columns:
    if data_new[i].dtypes=="object":
        data_new[i]=enc.fit_transform(data_new[i].values.reshape(-1,1))
```

###### Ensemble Technique

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
parameters={'criterion':['mse','mae'],'max_features':["auto","sqrt","log2"]}
rf=RandomForestRegressor()
clf=GridSearchCV(rf,parameters)
clf.fit(x_train,y_train)
print(clf.best_params_)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
import statsmodels.api as sm
rf=RandomForestRegressor(criterion="mse",max_features="log2")
rf.fit(x_train,y_train)
rf.score(x_train,y_train)
pred_decision=rf.predict(x_test)
rfs=r2_score(y_test,pred_decision)
print('R2 SCORE:',rfs*100)
rfscore=cross_val_score(rf,x,y,cv=5)
rfc=rfscore.mean()
print("Cross val score:",rfc*100)
```
